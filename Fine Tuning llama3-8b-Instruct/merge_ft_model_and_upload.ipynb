```python
from google.colab import userdata
HF_TOKEN = userdata.get('HF_TOKEN')
```

```python

OUTPUT_DIR = "/content/drive/MyDrive/llama-3b-instruct-ft-clean"
#MERGED_DIR = "/content/drive/MyDrive/fully_merged"
HF_REPO = "anonym-author/llama3-8b-instruct-arabic"
```

```python

print("Loading base model and LoRA adapters...")
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/llama-3-8b-Instruct",
    max_seq_length=2048,
    dtype=None,
    load_in_4bit=False,
)

from peft import PeftModel
model = PeftModel.from_pretrained(model, OUTPUT_DIR)
print("LoRA adapters loaded successfully!")

# -----------------------------
# Merge LoRA with Base Model
# -----------------------------
print("Merging LoRA adapters with base model...")
model = model.merge_and_unload()
print("Merge complete!")

print(f"Pushing to HuggingFace repo: {HF_REPO}...")
model.push_to_hub(HF_REPO, token=HF_TOKEN, private=False)
tokenizer.push_to_hub(HF_REPO, token=HF_TOKEN, private=False)
print(f"Successfully pushed to https://huggingface.co/{HF_REPO}")
```
